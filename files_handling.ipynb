##### 7. Count how many times the word "to" appears in the text file.

```python
file = r"C:\Users\zanca\OneDrive\Desktop\sample.txt"

with open(file)as text:
    file_str = text.read()

file_str_list = file_str.lower().split()

to_counter = 0

for word in file_str_list:
    if word == "to":
        to_counter += 1


print(to_counter)

```
