## Files handling.

##### 1. Open, read and print a txt file content.

```python
file = r"C:\Users\zanca\OneDrive\Desktop\sample.txt"

with open(file) as text:
    article = text.read()

print(article)

```

##### 2. Now, read line by line the same file.

```python
file = r"C:\Users\zanca\OneDrive\Desktop\sample.txt"

with open(file)as text:
    for lines in text:
        print(lines, end='')
```

##### 3. Read specific lines. Example: the first 5 lines.

```python
file = r"C:\Users\zanca\OneDrive\Desktop\sample.txt"

with open(file) as text:
    for i in range(5):
        print(text.readline(), end='')
```

##### 4. Counts the number of words in a text file.

```python
file = r"C:\Users\zanca\OneDrive\Desktop\sample.txt"

with open(file)as text:
    text_str = text.read()

text_list = text_str.lower().split()

print(len(text_list))
```
