##### 11. Create a file, write on it, then read the file and return an histogram of the words

```python
attempt = "Do you come from a land down under. Where women glow and men plunder? Can't you hear, can't you hear the thunder? You better run, you better take cover"
new_attempt_file = r"C:\Users\zanca\OneDrive\Desktop\new_attempt.txt"

def create_file(new_attempt_file, attempt):

    with open(new_attempt_file, "w") as file:

        return file.write(attempt)


def file_histogram(file) -> dict[str,int]:

    with open(file) as same_file:

        file_per_histo = same_file.read()

        list_file_per_histo = file_per_histo.lower().split()

        #print(type(list_file_per_histo))

        histogram = {}

        for word in list_file_per_histo:

            if word in histogram:

                histogram[word] += 1

            else:

                histogram[word] = 1

    return histogram


create_file(new_attempt_file, attempt)

print(file_histogram(new_attempt_file))    

```

##### 13. Create a function that takes a pattern string, a replacement string, and a  two filename as parameters. If the first file contains the pattern string, replace it with the replacement string and save the changes. Consider also to hnadle eventual errors, like file not found.

```python
# We need to create the "old" file. It will be made in order to have a "pattern".

pattern_file = r"C:\Users\zanca\OneDrive\Desktop\pattern_file.txt"

pattern = "* In the late days, * it was exceptionally rare * to spot an airplane flying over the sky, * and it was marvelous."

with open(pattern_file, "w") as file:

    file.write(pattern)



# We can now prceed with the actual task.

def sed(pattern: str, replacement: str, first_file: str, second_file: str) :

    try:

        with open(first_file, "r") as file_1:
            with open(second_file, "w") as renewed:

                source = file_1.read()

                for char in source:

                    source = char.replace(pattern, replacement)

                    renewed.write(source)
    
    except FileNotFoundError:
        print("The file was not found")

    except Exception as e:
        print(f"Unexpected error, {e}")  


sed("*", "", r"C:\Users\zanca\OneDrive\Desktop\pattern_fie.txt", r"C:\Users\zanca\OneDrive\Desktop\renewed_file.txt")    



```
