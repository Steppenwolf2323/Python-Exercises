## Object-oriented Programming in Python

##### 1. Create your fist Class with instance attributes.

```python
class Building:

    def __init__(self, name: str ,type: str, square: int):
        
        self.name = name
        self.type = type
        self.square = square

south_villa = Building("Villa Amarilla", "villa", 210)

print(south_villa.name)
print(south_villa.type)
print(south_villa.square)
```

##### 2. Create a sub_class that inherits the attributes of the parent class.

#

```python
class House(Building):
    pass

apartment_23 = House("Apartment 23", "Apartment", 56)

print(apartment_23.name)
print(apartment_23.type)
print(apartment_23.square)
```

##### 3. Create a new sub class, that inherits the attributes of the parent class but has its own unique attributes as well.

```python
class Houseboat(Building):

    def __init__(self, name:str, type:str, square:int, pp_fit:int):

        super().__init__(name,type,square)
        self.pp_fit = pp_fit

        

roosevelt = Houseboat("Roosvelt", "House Boat", 40, 2)

print(roosevelt.name,",",roosevelt.type,",",roosevelt.square,",",roosevelt.pp_fit)
```

##### 5. Add a new attribute to an existing subclass object. The new attribute should be calculated with a method inside the class itself. For example, the house both has an extra tax based on the square meters. The method should calculate the tax and add it as an attribute to the object.

```python
# First we add a new attribute to the subclass Houseboat
Houseboat.extra_tax = 0

# Then we create a method that it wil be added to the class Houseboat

def calculate_extra_tax(self):

    self.extra_tax = self.square * 0.15

# at this point we add the method to the subclass

Houseboat.calculate_extra_tax = calculate_extra_tax

roosevelt.calculate_extra_tax()
print(roosevelt.extra_tax)
```
