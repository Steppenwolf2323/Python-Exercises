##### 1. Create your fist Class with instance attributes.

```python
class Building:

    def __init__(self, name: str ,type: str, square: int):
        
        self.name = name
        self.type = type
        self.square = square

south_villa = Building("Villa Amarilla", "villa", 210)

print(south_villa.name)
print(south_villa.type)
print(south_villa.square)
```

##### 2. Create a sub_class that inherits the attributes of the parent class.

#

```python
class House(Building):
    pass

apartment_23 = House("Apartment 23", "Apartment", 56)

print(apartment_23.name)
print(apartment_23.type)
print(apartment_23.square)
```

##### 3. Create a new sub class, that inherits the attributes of the parent class but has its own unique attributes as well.

```python
class Houseboat(Building):

    def __init__(self, name:str, type:str, square:int, pp_fit:int):

        super().__init__(name,type,square)
        self.pp_fit = pp_fit

        

roosevelt = Houseboat("Roosvelt", "House Boat", 40, 2)

print(roosevelt.name,",",roosevelt.type,",",roosevelt.square,",",roosevelt.pp_fit)
```

##### 5. Add a new attribute to an existing subclass object. The new attribute should be calculated with a method inside the class itself. For example, the house both has an extra tax based on the square meters. The method should calculate the tax and add it as an attribute to the object.

```python
#Create a method that generates a new class attribute and it wil be added to the class Houseboat

def calculate_extra_tax(self):

    self.extra_tax = self.square * 0.15

# at this point we add the method to the subclass

Houseboat.calculate_extra_tax = calculate_extra_tax

roosevelt.calculate_extra_tax()
print(roosevelt.extra_tax)
```

##### 8. Create a class that includes subclasses for different shapes.Then create a method in the parent class, that calculates the area for each shape.

```python
class Shapes:

    def __init__(self, nr_sides: int):

        self.nr_sides = nr_sides


class Parallelogram(Shapes):

    def __init__(self, nr_sides:int, lenght_base:int, lenght_height:int):

        super().__init__(nr_sides)
        self.lenght_base = lenght_base
        self.lenght_height = lenght_height

class Circle(Shapes):

    def __init__(self, radius:int):

        super().__init__(nr_sides=0)
        self.radius = radius

class Triangle(Shapes):

    def __init__(self, nr_sides:int, lenght_base:int, lenght_height:int):
        
        super().__init__(nr_sides)
        self.lenght_base = lenght_base
        self.lenght_height = lenght_height


def area_calculator(self)-> str:

    if self.nr_sides == 3:
        area = (self.lenght_base * self.lenght_height) / 2
        return f"The area of the triangle is {area} square"
    
    elif self.nr_sides == 4:
        area = self.lenght_base * self.lenght_height
        return f"The area of the parallelogram is {area} square"

    else:
        area = 3.14 * (self.radius**2)
        return f"The area of the circle is {area} square"
    

Shapes.area_calculator = area_calculator



# Some testing

circle = Circle(4)
triangle = Triangle(3,4,6)
rectangle = Parallelogram(4,12,3)

print(circle.area_calculator())
print(triangle.area_calculator())
print(rectangle.area_calculator())        
```

##### 9. Create a class Student, with attributes like name, ID(random), and grades and average grade. Then create a class exams that inherits from student its ID and add the new grades from the exams to the student grades. The grades of the exams are input given by the user. Student also has a method that calculates the average grade from all the grades (initial grades + exam grades) and updates the average grade attribute.

```python
import random


class Student():

    def __init__ (self, name:str):

        self.name = name
        self.grades = []
        self.student_id = None
        self.average_grades = None

    
    def __str__(self):

        return f"Student name: {self.name}\nStudend ID:{self.student_id}\nGrades: {self.grades}\nGPA: {self.average_grades}\n"
    

    def generate_id(self):        
             
        self.student_id = random.randint(11111,99999)


    def enter_grades(self):
        
        grade = int(input(f"Please enter the result of exam "))
        self.grades.append(grade)
        


    def average(self):
                    
        self.average_grades = round((sum(self.grades))/(len(self.grades)),1)



selman = Student("Selman")
selman.generate_id()
selman.enter_grades()
selman.average()
print(selman)
selman.enter_grades()
selman.average()
print(selman)
selman.enter_grades()
selman.average()
print(selman)
selman.enter_grades()
selman.average()
print(selman)
selman.enter_grades()
selman.average()
print(selman)


#loete = Student("Loete")
#loete.generate_id()
#loete.enter_grades()
#loete.average()
#print(loete)

#karl = Student("Karl")
#karl.generate_id()
#print(karl)


```
