##### 8. Create a class that includes subclasses for different shapes.Then create a method in the parent class, that calculates the area for each shape.

```python
class Shapes:

    def __init__(self, nr_sides: int):

        self.nr_sides = nr_sides


class Parallelogram(Shapes):

    def __init__(self, nr_sides:int, lenght_base:int, lenght_height:int):

        super().__init__(nr_sides)
        self.lenght_base = lenght_base
        self.lenght_height = lenght_height

class Circle(Shapes):

    def __init__(self, radius:int):

        super().__init__(nr_sides=0)
        self.radius = radius

class Triangle(Shapes):

    def __init__(self, nr_sides:int, lenght_base:int, lenght_height:int):
        
        super().__init__(nr_sides)
        self.lenght_base = lenght_base
        self.lenght_height = lenght_height


def area_calculator(self)-> str:

    if self.nr_sides == 3:
        area = (self.lenght_base * self.lenght_height) / 2
        return f"The area of the triangle is {area} square"
    
    elif self.nr_sides == 4:
        area = self.lenght_base * self.lenght_height
        return f"The area of the parallelogram is {area} square"

    else:
        area = 3.14 * (self.radius**2)
        return f"The area of the circle is {area} square"
    

Shapes.area_calculator = area_calculator



# Some testing

circle = Circle(4)
triangle = Triangle(3,4,6)
rectangle = Parallelogram(4,12,3)

print(circle.area_calculator())
print(triangle.area_calculator())
print(rectangle.area_calculator())        
```

##### 11. Create a Bank Account class, with attributes like account number, balance, and account holder name. Then create methods to deposit, withdraw, and check the balance.  You also need to be able to calculate fixed interest on the balance and fixed fees (they are constant values linked to the class).

```python
class BankAccount:

    fee_amount = 30
    interest_rate = 0.2

    def __init__(self, name: str, max_withdraw: float, balance: float = 0)-> None:

        self.name = name        
        self.max_withdraw = max_withdraw
        self.balance = balance


    def deposit(self, amount: float)-> None:

        if amount > 0:
            self.balance += amount
        else:
            print("You can't deposit a negative amount.")


    def withdraw(self, amount: float) -> None:

        if amount <= self.balance and amount <= self.max_withdraw:
            self.balance -= amount
        else:
            print(f"Withdraw failed: check your balance or the 'withdraw' limit.")


    def apply_fee(self) -> None:

        self.balance -= self.fee_amount


    def apply_interest(self) -> None:

        self.balance += self.balance * self.interest_rate

    
    def get_balance(self) -> float:

        return self.balance
    
    
    def __str__(self) -> str:

        return f"Account of {self.name}\n Balance: {self.balance}\n Interest rate: {self.interest_rate}\n Bank Fee: {self.fee_amount}"
    


alice = BankAccount("Alice", 500) 
bob = BankAccount("Bob", 100)

alice.deposit(200)
bob.deposit(300)
alice.withdraw(50)

alice.apply_interest()
bob.apply_fee()

print(alice.get_balance())
print(bob.get_balance())

print(alice)

bob.withdraw(200)
alice.withdraw(200)
alice.deposit(-10)        
    
```

##### 12. Rock, Paper, Scissors Game: Create a Player Class and a Move Class. 

```python
class Player:

    def __init__(self, name: str, score: int = 0) -> None:
        self.name = name
        self.score = score

    def __str__(self):
        return f"{self.name} score: {self.score}"


class Move:

    def __init__(self, choice: str) -> None:
        self.choice = choice.lower()

    def game_logic(self, player_1: Player, p1_move, player_2: Player, p2_move):

        if p1_move.choice == p2_move.choice:
            print("It's a tie!")
            return
        
        if p1_move.choice == "paper" and p2_move.choice == "rock":
            player_1.score += 1
            print(f"{player_1.name} wins!")
        
        elif p1_move.choice == "rock" and p2_move.choice == "scissors":
            player_1.score += 1
            print(f"{player_1.name} wins!")

        elif p1_move.choice == "scissors" and p2_move.choice == "paper":
            player_1.score += 1
            print(f"{player_1.name} wins!")

        else:
            player_2.score += 1
            print(f"{player_2.name} wins!")


anna = Player("Anna")
bill = Player("Bill")

anna_move = Move(input("Anna, choose 'rock', 'paper' or 'scissor'"))
bill_move = Move(input("Bill, choose 'rock', 'paper' or 'scissor'"))

anna_move.game_logic(anna, anna_move, bill, bill_move)

print(anna)
print(bill)

```

##### 13. Create a chess board with the following specifications:
- The board should be represented as an 8x8 grid.
- A method should check if a move is valid based on the piece's movement rules.
- A method that check if a square is occupied or not.

```python
class Board:

    def __init__(self) -> None:
        self.grid = []
        for _ in range(8):              
            row = []
            for _ in range(8):          
                row.append(None)       
            self.grid.append(row)

    def in_bounds(self, row: int, col: int) -> bool:

        return 0 <= row < 8 and 0 <= col < 8
    
    def is_occupied(self, row:int, col: int) -> bool:

        return self.in_bounds(row,col) and self.grid[row][col]
```
