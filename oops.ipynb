##### 1. Create your fist Class with instance attributes.

```python
class Building:

    def __init__(self, name: str ,type: str, square: int):
        
        self.name = name
        self.type = type
        self.square = square

south_villa = Building("Villa Amarilla", "villa", 210)

print(south_villa.name)
print(south_villa.type)
print(south_villa.square)
```

##### 8. Create a class that includes subclasses for different shapes.Then create a method in the parent class, that calculates the area for each shape.

```python
class Shapes:

    def __init__(self, nr_sides: int):

        self.nr_sides = nr_sides


class Parallelogram(Shapes):

    def __init__(self, nr_sides:int, lenght_base:int, lenght_height:int):

        super().__init__(nr_sides)
        self.lenght_base = lenght_base
        self.lenght_height = lenght_height

class Circle(Shapes):

    def __init__(self, radius:int):

        super().__init__(nr_sides=0)
        self.radius = radius

class Triangle(Shapes):

    def __init__(self, nr_sides:int, lenght_base:int, lenght_height:int):
        
        super().__init__(nr_sides)
        self.lenght_base = lenght_base
        self.lenght_height = lenght_height


def area_calculator(self)-> str:

    if self.nr_sides == 3:
        area = (self.lenght_base * self.lenght_height) / 2
        return f"The area of the triangle is {area} square"
    
    elif self.nr_sides == 4:
        area = self.lenght_base * self.lenght_height
        return f"The area of the parallelogram is {area} square"

    else:
        area = 3.14 * (self.radius**2)
        return f"The area of the circle is {area} square"
    

Shapes.area_calculator = area_calculator



# Some testing

circle = Circle(4)
triangle = Triangle(3,4,6)
rectangle = Parallelogram(4,12,3)

print(circle.area_calculator())
print(triangle.area_calculator())
print(rectangle.area_calculator())        
```

Student also has a method that calculates the average grade from all the grades (initial grades + exam grades) and updates the average grade attribute.

```python
import random


class Student():

    def __init__ (self, name:str):

        self.name = name
        self.grades = []
        self.student_id = None
        self.average_grades = None

    
    def __str__(self):

        return f"Student name: {self.name}\nStudend ID:{self.student_id}\nGrades: {self.grades}\nGPA: {self.average_grades}\n"
    

    def generate_id(self):        
             
        self.student_id = random.randint(11111,99999)


    def enter_grades(self):

        try:         
            grade = int(input(f"Please enter the result of exam "))
        except ValueError:
            print("The grade must to be a number")



        if grade < 0:
            print("The grade must to be a positive integer.")
        elif grade > 10:
            print("The grade cannot be bigger than 10.")
        else:        
            self.grades.append(grade)
        


    def average(self):

        if len(self.grades) == 0:
            self.average_grades = None
        else:                    
            self.average_grades = round((sum(self.grades))/(len(self.grades)),1)



```

##### 11. Create a Bank Account class, with attributes like account number, balance, and account holder name. Then create methods to deposit, withdraw, and check the balance.  You also need to be able to calculate fixed interest on the balance and fixed fees (they are constant values linked to the class).

```python
class BankAccount:

    fee_amount = 30
    interest_rate = 0.2

    def __init__(self, name: str, balance: float, max_withdraw: float)-> None:

        self.name = name
        self.balance = balance
        self.max_withdraw = max_withdraw


    def deposit(self, amount: float)-> None:

        if amount > 0:
            self.balance += amount
        else:
            print("The amount entered must to be bigger than 0")


    def withdraw(self, amount: float) -> None:

        if amount <= self.balance and amount <= self.max_withdraw:
            self.balance -= amount
        else:
            print(f"Withdraw failed: check your balance or the 'withdraw' limit.")


    def apply_fee(self) -> None:

        self.balance -= self.fee_amount


    def apply_interest(self) -> None:

        self.balance *= self.interest_rate

    
    def get_balance(self) -> float:

        return self.balance
    
    
    def __str__(self) -> str:

        return f"Account on behalfe of {self.name}\n Balance: {self.balance}\n Interest rate: {self.interest_rate}\n Bank Fee: {self.fee_amount}"
    


    

    


        
    
```
