##### 2. Check if a list contains only numbers. If it contains any non-numeric values, handle the exception and print an appropriate message.

```python
def check_list_numbers(list_char)->None:

    for index in range(len(list_char)):
        try:        
            if isinstance(list_char[index],int):
                print(f"At index {index} we do have an integer: {list_char[index]}")
            elif isinstance(list_char[index],float):
                print(f"At index {index} we do have a float: {list_char[index]}")
            else:
                raise ValueError
        
        except ValueError:
            print(f"At index {index} there is '{list_char[index]}', which is not a number ")


check_list_numbers([10, 20, 'rebel', 40, 4.2])
print(f"")
check_list_numbers([44, 0.32, 78, 91, 1212.34])
```

##### 3. Create a function that checks if a given dish is gluten-free. If the food item does not specify whether it is gluten-free, raise a custom exception and handle it appropriately.

```python
def gluten_free_check(dishes: list[tuple])->None:

    for dish in dishes:   
    
        try:
        
            if dish[1] is True:
                print(f"{dish[0]} --> gluten")     
                   
            elif dish[1] is False:
                print(f"{dish[0]} --> gluten free")

            else:
                raise Exception
                         
            
        except Exception:
            print(f"{dish} --> Information not available")         
   

dishes = [("Ceaser Salad", True), ("Carbonara", False), ("Welsh Rarebit"), 
          ("Croque monsieur", True), ("Meat_balls", True), ("Pizza Margherita", False) ,
            ("Pasta al Pesto", True) , ("Sate' Ayam"), ("Tacos", False)]        


gluten_free_check(dishes)
```

##### 4. Implement a function that takes a list of strings and converts them to integers. In case of errors, handle the exceptions and print which strings could not be converted.

```python
def convert_integers(numbers: list[str])-> list[int]:

    list_integers = []

    for item in numbers:

        try:
            num = int(item)
            list_integers.append(num)
        except ValueError:
            print(f"'{item}' cannot be converted")

    return(f"The list of integers is: {list_integers}")

list_char = ['10', '0.5', 'abc', '20', 'xyz', '30', 'two']

convert_integers(list_char)
```

##### 6. Quality check: a function takes a car and its specifications as input. If the specifications are not meet, raise a custom exception. Also, the specifications has to be checked for value errors.


```python
# Our car need to be white or grey(type str), 4 doors(type int), 2.0 engine(type float) or bigger till 3.0, and German(type str)

def check_car(car: list):    

    # check type value

    try:
        if not isinstance(car[0],str):
            raise ValueError
    except ValueError:
        print(f"ERROR: 'Color' value has to be 'str'")

    try:
        if not isinstance(car[1], int):
            raise ValueError
    except ValueError:
        print(f"ERROR: 'Nr_doors' value has to be 'int'")

    try:
        if not isinstance(car[2],float):
            raise ValueError
    except ValueError:
        print(f"ERROR: 'Engine' value has to be float")

    try:
        if not isinstance(car[3], str):
            raise ValueError
    except ValueError:
        print((f"ERROR: 'Nation' value has to be 'str'"))
        
      
    
    # check values
    
    if car[0] == "white" or car[0] == "grey":
            print("The color is accepted")
    else:
        print(f"The color {car[0]} is not accepted.")
    
    if car[1] == 4:
        print("The number of doors is accepted")
    else:
        print(f"The number of doors {car[1]} is not accepted.")

    if 2.0 <= car[2] <= 3.0:
        print("The engine is accepted")
    else:
        print(f"The engine {car[2]} is not accepted.")

    if car[3] == "Germany":
        print("The Nation specification is accepted")
    else:
        print(f"The Nation of the car, {car[3]}, is not accepted.")


car_1 = ["white", 4, 2.5, "Japan"]
car_2 = ["red", "four", 3.0, "Germany"]
car_3 = ["grey", 4, 20, 89]
car_4 = ["grey", 2, 4.0, "Germany"]       

check_car(car_1)
print("\n")
check_car(car_2)
print("\n")
check_car(car_3)
print("\n")
check_car(car_4)
print("\n")       
            
```

##### 7 In a dictionary, check if the value is a str. Then check if that value is a mix of letters and numbers. Then check if the length of the string is exactly 8 characters.

```python
sample_dict = {
    "user1": "abc12345",  
    "user2": "ABCDEFGH",   
    "user3": "12345678",   
    "user4": "a1b2c3",     
    "user5": 98765432,    
    "user6": "ab12cd34",   
    "user7": "hello",      
    "user8": "xy9z8w7q",   
}

def password_checks(sample_dict) -> None:


    # 1. Check that the password is a string type

    try:

        for password in sample_dict.values():

            if not isinstance(password, str):

                raise ValueError
            
            else:

                print("The password is a str.")

    except ValueError:

        print(f"Not all password are string.\n")    


    # 2. Check if there are at least one letter or one number

    try:        

        for password in sample_dict.values():

            letters = False
            numbers = False

            for char in password:

                if char.isalpha():

                    letters = True

                if char.isdigit():

                    numbers = True

            if letters and numbers:

                print("The password has letters and numbers.")

            else:

                raise ValueError
        
    except ValueError:

        print(f"The password must contain at least one letter and one number.\n")

    
    # 3. Check the lenght of the string

    try:

        for password in sample_dict.values():

            if len(password)!= 8:

                raise ValueError
            
            else:

                print ("The password has the exact lenght.")

    except ValueError:

        print(f"The password does not have the right lenght.\n")


password_checks(sample_dict)
```

##### 8. Create a funcion to check a single password. The function should check different conditions, and raise exceptions for each failed condition, so that the code does not stop at the first failure. The conditions are:
- lenght of password
- contains at least one uppercase letter
- contains at least one special character
- no spaces allowed

##### The conditions (as the password to be check) are given as parameters to the function, so that eventually not all conditions have to be checked.

```python
def password_check(password: str, checks: tuple[int,bool, bool, bool]) -> str:

    special_char = "!@#$%^&*(),./?:;<>"

    error_message = []

    # 1 Check lenght

    if checks[0] != 0 and len(password) != checks[0]:
        error_message.append(f"The password has to be {checks[0]} characters long. ")

    # 2. Check uppercase

    if checks[1]:

        upper = False

        for char in password:

            if char.isupper():

                upper = True

        if not upper:

            error_message.append(f"The password need at least one uppercase character.")

    # 3. Check special case

    if checks[2]:

        special = False

        for char in password:

            if char in special_char:

                special = True

        if not special:

            error_message.append(f"The password need at least one special character.")

    # 4. Check for spaces

    if checks[3] and ' ' in password:
        error_message.append("The password have at least one space. That's not allowed.")

    # 5. Raise the error and display the message

    if error_message:
            raise Exception("\n".join(error_message))
    else:
        print("Password accepted")


        
password = "reat password"
password_checks = (12, True, True, True)

password_check(password, password_checks)
```
