##### 1. Ask the user to enter two numbers, then perform integer division and print the result. Handle any exceptions that may occur.

```python
try:
    a = float(input("Enter a number."))
    b = float(input("Enter a number."))

    c = round((a / b) , 3)

    print(f"The result of the division is {c}")

except ValueError:
    print("You have entered a different character than a number.")

except ZeroDivisionError:
    print("You cannot divide by zero.")
```

##### 2. Check if a list contains only numbers. If it contains any non-numeric values, handle the exception and print an appropriate message.

```python
def check_list_numbers(list_char)->None:

    for index in range(len(list_char)):
        try:        
            if isinstance(list_char[index],int):
                print(f"At index {index} we do have an integer: {list_char[index]}")
            elif isinstance(list_char[index],float):
                print(f"At index {index} we do have a float: {list_char[index]}")
            else:
                raise ValueError
        
        except ValueError:
            print(f"At index {index} there is '{list_char[index]}', which is not a number ")


check_list_numbers([10, 20, 'rebel', 40, 4.2])
print(f"\n")
check_list_numbers([44, 0.32, 78, 91, 1212.34])
```

##### 3. Create a function that checks if a given dish is gluten-free. If the food item does not specify whether it is gluten-free, raise a custom exception and handle it appropriately.

```python
dishes = [("Ceaser Salad", True), ("Carbonara", False), ("Welsh Rarebit"), 
          ("Croque monsieur", True), ("Meat_balls", True), ("Pizza Margherita", False) ,
            ("Pasta al Pesto", True) , ("Sate' Ayam"), ("Tacos", False)]



def gluten_free_check(dishes: list[tuple]):

    for dish in dishes:   
    
        try:
        
            if dish[1] is True:
                print(f"{dish[0]} --> gluten")     
                   
            elif dish[1] is False:
                print(f"{dish[0]} --> gluten free")

            else:
                raise Exception
                         
            
        except Exception:
            print(f"{dish} --> Information not available")         
   
        


gluten_free_check(dishes)
```
