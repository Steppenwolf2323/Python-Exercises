## Exception in Python

##### 1. Ask the user to enter two numbers, then perform integer division and print the result. Handle any exceptions that may occur.

```python
try:
    a = float(input("Enter a number."))
    b = float(input("Enter a number."))

    c = round((a / b) , 3)

    print(f"The result of the division is {c}")

except ValueError:
    print("You have entered a different character than a number.")

except ZeroDivisionError:
    print("You cannot divide by zero.")
    
```
