##### 1. Manually create a dictionary, then perform the following basic operations:

##### 6. From a string, the letters become keys and their frequency becomes values in a dictionary.

```python
long_string = "And they said it was fine to be broken"

frequency_dict= {}

for letter in long_string:
    if letter.lower() not in frequency_dict:
        frequency_dict[letter.lower()] = 1
    else:
        frequency_dict[letter.lower()] += 1

print(frequency_dict)
```

##### 11. Implement a reverse lookup function that takes a dictionary and a value, and returns the first key that corresponds to that value.

```python
my_dict = {'a': 1, 'b': 2, 'c': 3, 'd': 2}
val_1 = 3
val_2 = 5

def reverse_lookup(straight:dict, val:int):
    
    for key in straight:
        if straight[key] == val:
            return f"'{key}' is the first value of {val} found."
    raise LookupError(f">>> Value of '{val}' not found <<<")
    
print(reverse_lookup(my_dict,val_1))
print(reverse_lookup(my_dict,val_2))

```

##### 12. A dictionary have for different keys, same values. Create a new dictionary that groups keys by their values.

```python
def invert_key_value(base_dict: dict)-> dict:

    value_key_dict = {}
    for key in base_dict:        
        if base_dict[key] not in value_key_dict:
            value_key_dict[base_dict[key]] = [key]
        else:
            value_key_dict[base_dict[key]].append(key)

    return value_key_dict

base_dict_1 = {'apple': 20, 'banana': 30, 'cherry': 20, 'date': 40, 'elderberry': 30, 'fig': 50, 'grape': 40, 'honeydew': 60, 'kiwi': 50, 'lemon': 20, 'mango': 60, 'nectarine': 30, 'orange': 50, 'papaya': 20, 'quince': 50}
inventory = invert_key_value(base_dict_1)
print(f"Fruits in the storage:\n{inventory}")
print(f"\n")

base_dict_2 = { "Adams": "Client", "Frank": "Director", "Peter": "Manager", "Nichols": "Supplier", "Owen": "Supervisor", "Bill" : "Supervisor", "Baker": "Client", "Carter": "Supplier", "Davis": "Client", "Evans": "Supplier"}
draft_database = invert_key_value(base_dict_2)
print(f"About the shop:\n{draft_database}")
```

##### 13. A function open a text file and counts the frequency of each word, storing the results in a dictionary.

```python
def words_frequency(text:str)-> dict:

    with open(text,"r") as file:
        list_words = file.read().split()

    histogram = {}

    for word in list_words:
        if word not in histogram:
            histogram[word] = 1
        else:
            histogram[word] += 1
    
    return histogram



#word_dict = words_frequency(r"C:\Users\zanca\OneDrive\Desktop\sample.txt")
#for item in word_dict.items():
#    print(item)
```

##### 14. Concatenate two dictionaries, summing the values of common keys. Find the key with the highest value in the resulting dictionary.


```python
def concatenate_dict(dict_1,dict_2) -> dict:

    new_dict ={}

    for item in dict_1:
        new_dict[item] = dict_1[item]
    
    for pair in dict_2:
        if pair not in new_dict:
            new_dict[pair] = dict_2[pair]
        else:
            new_dict[pair] += dict_2[pair]
    
    sorted_new_dict = sorted(new_dict.items(), key=lambda x: x[1], reverse=True)

    return sorted_new_dict

east = {'a': 2, 'b': 3, 'c': 1}
west = {'a': 4, 'c': 5, 'e': 2} 


res = concatenate_dict(east,west)
for elements in res:
    print(f"Region 'EW', type '{elements[0].upper()}' has a total of {elements[1]} units")
```
