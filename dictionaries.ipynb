## Dictionaries in Python

##### 1. Manually create a dictionary, then perform the following basic operations:
- Add a new key-value pair
- Update the value of an existing key
- Print  the value of a specific key

```python
student_datas = {"name" : "Robert", "age" : 17, "city" : "Dublin"}
print ("The initial dictionary: ", student_datas)

# adding a couple of new key-value pairs

student_datas["working student"] = False
student_datas["sport"] = "basketball"
print("The second version of the dictionary:" , student_datas)

# updating a couple of values 

student_datas["age"] = 18
student_datas["working student"] = True
print("The third version of the dictionary:", student_datas)

# printing a couple of values

print(student_datas["name"], student_datas["age"])
```
