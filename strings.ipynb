## String in Python

##### 1. Out of a string, create a new string made of only vowels. 


```python
surname = "Callagher"

vowels ="AEIOUaeiou"

vowels_surname = ''

for letter in surname:
    if letter in vowels:
        vowels_surname += letter

print(vowels_surname)
```

##### 2. Given a string, create a new string made of the first, the middle, and the last character.

```python
short_sentence = "It is definetely a beautiful day"

index_mid_char = int(len(short_sentence)/2)

first_last_mid = short_sentence[0] + ' ' + short_sentence[index_mid_char-1] + ' ' + short_sentence[-1]

print(first_last_mid)
```

##### 3. From a given string, display only the 3 middle characters.

```python
long_string = "Supercalifragilisticexpialidocious"

mid_char = int(len(long_string)/2)

print(long_string[mid_char-2]+long_string[mid_char-1]+long_string[mid_char])
```

##### 4. From a given string, display oonly the characters that are present at an even index.

```python
given_string = "Eventually, check the weather before to leave"
even_index_letter = ""

for letter in range(len(given_string)):
    if letter % 2 == 0:
        even_index_letter += given_string[letter]

print(even_index_letter)
```

##### 5. Given two strings, a new string is made from the second sentence enterded the middle of the first string. Example: "HelloWorld" and "Python" → "HelloPythonWorld".

```python
sentence_1 = "HelloWorld"
sentence_2 = "Python"

new_sentence = ""

mid = int(len(sentence_1) / 2)

new_sentence = sentence_1[:mid] + sentence_2 + sentence_1[mid:]

print(new_sentence)
```

##### 6. Create a new string made of the even index character in the first string and the odd index characters from the second string. Choose two strings with the same length. Example: "Giraffe" and "Leopard" → "Gerpfre"

```python
animal_1 = "Giraffedd"
animal_2 = "Leopard"

even_from_1 = ""
odd_from_2 = ""

final = ""



for letter in range(len(animal_1)):
    if letter % 2 == 0:
        even_from_1 += animal_1[letter]

for letter in range(len(animal_2)):
    if letter % 2 != 0:
        odd_from_2 += animal_2[letter] 

for x,j in zip(even_from_1,odd_from_2):
    final += x + j
   
print(final)
```
