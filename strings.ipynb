## String in Python

##### 1. Out of a string, create a new string made of only vowels. 


```python
surname = "Callagher"

vowels ="AEIOUaeiou"

vowels_surname = ''

for letter in surname:
    if letter in vowels:
        vowels_surname += letter

print(vowels_surname)
```

##### 2. Given a string, create a new string made of the first, the middle, and the last character.

```python
short_sentence = "It is definetely a beautiful day"

index_mid_char = int(len(short_sentence)/2)

first_last_mid = short_sentence[0] + ' ' + short_sentence[index_mid_char-1] + ' ' + short_sentence[-1]

print(first_last_mid)
```

##### 3. From a given string, display only the 3 middle characters.

```python
long_string = "Supercalifragilisticexpialidocious"

mid_char = int(len(long_string)/2)

print(long_string[mid_char-2]+long_string[mid_char-1]+long_string[mid_char])
```

##### 4. From a given string, display oonly the characters that are present at an even index.

```python
given_string = "Eventually, check the weather before to leave"
even_index_letter = ""

for letter in range(len(given_string)):
    if letter % 2 == 0:
        even_index_letter += given_string[letter]

print(even_index_letter)
```
