##### 1. Out of a string, create a new string made of only vowels. 


```python
surname = "Callagher"

vowels ="AEIOUaeiou"

vowels_surname = ''

for letter in surname:
    if letter in vowels:
        vowels_surname += letter

print(vowels_surname)
```

##### 5. Given two strings, a new string is made from the second sentence enterded the middle of the first string. Example: "HelloWorld" and "Python" → "HelloPythonWorld".

##### 6. Create a new string made of the even index character in the first string and the odd index characters from the second string. Choose two strings with the same length. Example: "Giraffe" and "Leopard" → "Gerpfre"

```python
animal_1 = "Giraffedd"
animal_2 = "Leopard"

even_from_1 = ""
odd_from_2 = ""

final = ""



for letter in range(len(animal_1)):
    if letter % 2 == 0:
        even_from_1 += animal_1[letter]

for letter in range(len(animal_2)):
    if letter % 2 != 0:
        odd_from_2 += animal_2[letter] 

for x,j in zip(even_from_1,odd_from_2):
    final += x + j
   
print(final)
```

##### 7. Riarrange the characters of a string in a way that the lowercase letters should come first. Example: "DC Comics" → "omicsDCC"

```python
inp_string = "DC Comics"

lower_case = ""
upper_case = ""

riarranged = ""

for letter in inp_string:
    if letter.islower():
        lower_case += letter
    elif letter.isupper():
        upper_case += letter
    else:
        continue

riarranged = lower_case + upper_case

print(riarranged)

```

##### 8. From a given random string, divided the different element of the string, based on the character type.

```python
rand_string = "Gn34@n90-1q##_25de"

letters = ""
digits = ""
chars = ""

for x in rand_string:
    if x.isalpha():
        letters += x
    elif x.isdigit():
        digits += x
    else:
        chars += x

print(letters)
print(digits)
print(chars)
```

##### 9. You have first_string = "Abc" and second_string = "Xyz". Create third_string = "AzbycX"

##### 11. Find how many times a word occured in a string.

```python
sentence = "The table in the kitchen and the piano in the living room are piece of the old owner of the house."

# Find how many time "the" occur.
searched = "in"

finding_the = sentence.lower().count(searched.lower())

print(f"The word '{searched}' occured {finding_the} times.")
```

```python
sentence = "The table in the kitchen and the piano in the living room are piece of the old owner of the house."

# Find how many time "the" occur.
searched = "in"
counter = 0
words_in_sentence = sentence.lower().split()

for word in words_in_sentence:
    if word == searched:
        counter += 1

print(f"The word '{searched}' occured {counter} times.")
```

##### 12. Make the sum of the digits present in a string, and their average.

```python
string_numbers = "132ds4f4w2e3F"

total = 0
counter = 0

for x in string_numbers:
    if x.isdigit():
        total += int(x)
        counter += 1

avg = total / counter

print(f"The total of the digit in the sentence is {total}.\nThe average of the sum of the numbers in the sentence is {round(avg,2)}.")
```

##### 14. Reverse a given string.

```python
straight_string = "A green apple."

reversed_string = ""

for char in range(len(straight_string),0,-1):
    reversed_string += straight_string[char-1]

print(reversed_string)
```

```python
straight_string = "A green apple."

reversed_string = "".join(reversed(straight_string))

print(reversed_string)
```

##### 19. Count vowels in a string still usign .count()

```python
vowels = 'aeiou'
sentence = 'The day after tomorrow is a popular movie of...'

amount_vowels = 0

for char in vowels:
    if char.lower() in sentence:
        amount_vowels += sentence.count(char)
        print(char, ":" , sentence.count(char))

print("Total vowels:", amount_vowels)
```
